```python
#check guess if valid
def checkGuess(guess):    
        
    checkGuess.goodGuess = False

    #while True:
        #get guess
            
        #retsart when guess invalid
    if guess not in allwords:
        invalidBar()
        #break
            
        #if guess valid check through checker
    if guess in allwords:
        checkGuess.goodGuess = True
        
        try:
            checker(guess, newword)

        except NameError:
            checker(guess, answer)

        #switch round number
        def SWITCH(rdnum):
            SWITCH.truefalse = True

            SWITCH.roundeditedval = rdnum
            if SWITCH.truefalse == True:
                SWITCH.roundeditedval = SWITCH.roundeditedval + 1
                SWITCH.truefalse = False
            #print(SWITCH.roundeditedval)

        #set variables for boxes and round number        
        SWITCH(definegm.roundsplayed)

        #check if right guess
        try:
            if guess == newword:
                OverRightGuess()
        
        except NameError:
            if guess == answer:
                OverRightGuess()

        #check if game over
        if definegm.roundsplayed == 6:
            
            try:
                if guess != newword:
                    OverNoGuess()

            except NameError:
                if guess != answer:
                    OverNoGuess()

        definegm.roundsplayed = SWITCH.roundeditedval
        #break
```

```python
#delete button logic
def deleteButton():
    
    entryvar2 = entrybox.get()
    
    #make into all caps
    if entryvar2.isupper() == False:
      entryvar2 = entryvar2.upper()

    #delete from entrybox before pressing enter
    if int(len(entrybox.get())) > 0:

        entrybox.delete(  (len(entrybox.get()) - 1), len(entrybox.get())   )  
        return

    #check if entry is greater then 0
    if len(entryvar2) == 0:
        return
    
    #delete after pressing enter if entry is > 0
    if len(entryvar2) > 0:
        
        #change entrybox ONLY when entry != label
        if str(entryvar2.upper()) != str(entrybox.get()):
            entrybox.delete(  (len(entrybox.get()) - 1), len(entrybox.get())   ) 
            return

        #change label and entry when entry = label
        if str(entrybox.get()) == str(entryvar2.upper()):
            
            croppedvar = entrybox.get()[:-1]
            
            #txtlabel.config(text = croppedvar)
            #displayText()
            return
        
        return

    #break out when no string
    if len(croppedvar) == 0:
        #txtlabel.config(text = croppedvar)
        #displayText()
        return
    pass
```
